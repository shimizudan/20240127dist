## 逆関数法（乱数の生成）
命題2.18のところで，「乱数を形成するのに使われる（逆関数法）」とあったので，Julia言語で実装してみました。

```julia
using Distributions,Plots# 分布パッケージ，プロットパッケージ

r = rand(10000) # 区間[0,1]で一様分布に従う乱数（一様乱数）を10,000個作る。

f(x) = quantile(Normal(),x) # 標準正規分布(Normal())の分布関数の逆関数(quantile）

g(x) = quantile(Binomial(100,0.97),x) # 二項分布(Normal(100,.97))の分布関数の逆関数(quantile）

h(x) = quantile(Exponential() ,x) # 指数分布(Exponential(1))の分布関数の逆関数(quantile）

```

```julia
histogram(r,label="Uniform()") #一様乱数をヒストグラムで確認
```

```julia
f.(r) #標準正規分布(Normal())に従う乱数（正規乱数）10,000個
```

```julia
histogram(f.(r),label="Normal(0,1)") # ヒストグラムで確認
```

```julia
g.(r) #二項分布(Normal(100,.97))に従う乱数（正規乱数）10,000個
```

```julia
histogram(g.(r),label="Binomial(100,0.97)")
```

```julia
h.(r) #指数分布(Exponential(1))に従う乱数10,000個
```

```julia
histogram(h.(r),label="Exponential(1)")
```
